```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

spotify_df = pd.read_csv("../data/Cleaned_Spotify_2024_Global_Streaming_Data.csv")
spotify_df.head()
```

```python
avg_streams = spotify_df.groupby('Genre')['Total Streams (Millions)'].mean().sort_values(ascending=False).head(10)
```

```python
plt.Figure(figsize=(10,5))
plt.bar(avg_streams.index,avg_streams.values)
plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

```python

plt.Figure(figsize=(10,5))
plt.bar(avg_streams.index,avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel("Genre")
plt.ylabel("Average Total Streams (Millions)")
plt.title("Average Total Streams by Genre")
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig("../plots/Average_Streams_by_Genre.png")
plt.show
```
